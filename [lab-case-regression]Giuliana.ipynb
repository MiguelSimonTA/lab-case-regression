```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')
```

```python
categorical = pd.read_csv('/Users/giulianamiranda/Documents/Labs/lab-case-regression/categorical.csv') 
categorical.describe(include='all')
```

```python
numerical = pd.read_csv('/Users/giulianamiranda/Documents/Labs/lab-case-regression/numerical.csv') 
numerical.describe(include= 'all')
```

```python
target = pd.read_csv('/Users/giulianamiranda/Documents/Labs/lab-case-regression/target.csv') 
target.describe(include= 'all')
```

```python
# To begin with, I'm creating a filter to select only the data on the people who actually donated

full_df = pd.concat([numerical, categorical], axis=1) 
```

```python
full_df = pd.concat([full_df, target], axis=1) 
```

```python
df_contributed = full_df[full_df['TARGET_B'] == 1]
```

```python
X = df_contributed.drop(['TARGET_B', 'TARGET_D'], axis = 1)
```

```python
y = df_contributed[['TARGET_B', 'TARGET_D']]
```

```python
X_num = X.select_dtypes(include='number')
X_cat = X.select_dtypes(include='object')
```

```python
X_cat = X_cat.astype(str)

encoder = OneHotEncoder(drop='first').fit(X_cat)
encoded_categorical_array = encoder.transform(X_cat).toarray()
encoded_categorical = pd.DataFrame(encoded_categorical_array, columns=encoder.get_feature_names_out(X_cat.columns))


```

```python
scaler = StandardScaler()
numericalX_norm = scaler.fit_transform(X_num)
```

```python
numericalX_norm = pd.DataFrame(numericalX_norm, columns=X_num.columns)
```

```python
# I'm reseting the index to prevent erros
numericalX_norm = numericalX_norm.reset_index(drop = True)
encoded_categorical = encoded_categorical.reset_index(drop = True)
```

```python
X_new = pd.concat([numericalX_norm, encoded_categorical], axis=1)
y = y = df_contributed[['TARGET_D']]
```

```python
# Doing the train test split

X_train, X_test, y_train, y_test = train_test_split(X_new, y, random_state = 0, test_size = 0.2)

```

```python
from sklearn.linear_model import LinearRegression

lin_r = LinearRegression()
lin_r.fit(X_train, y_train)
y_pred = lin_r.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error (MSE): {mse}')

r_squared = r2_score(y_test, y_pred)
print(f'R-squared: {r_squared}')
```
