```python
# Importing libraries
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn import linear_model
```

```python
# Importing data
donors = pd.read_csv(r"C:\Users\joaoa\Desktop\Ironhack\Downloads_Ironhack\learningSet.txt")
donors = donors[donors["TARGET_B"] == 1]
donors
```

```python
# Creating a new dataframe with numerical data and dropping null values and TARGET_B 
# I decided to perform the model with only numerical data since there are already hundreds of numerical variables
numerical = donors.select_dtypes(include=np.number)
numerical = numerical.dropna(axis = 1)
numerical = numerical.drop("TARGET_B", axis = 1)
numerical
```

```python
from sklearn.model_selection import train_test_split

# Defining X and y variables
X = numerical.drop("TARGET_D",axis = 1)
y = numerical["TARGET_D"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.2)

# Train dataframe
df_train = pd.concat([X_train, y_train], axis = 1)
df_train = df_train.reset_index(drop = True)
df_train
```

```python
# Test dataframe
df_test = pd.concat([X_test, y_test], axis = 1)
df_test = df_test.reset_index(drop = True)
df_test
```

```python
from sklearn.preprocessing import StandardScaler

# Standardizing X_train variables
scaler = StandardScaler()
X_train_stand = scaler.fit_transform(df_train.drop("TARGET_D", axis = 1))
X_train_stand = pd.DataFrame(X_train_stand, columns = df_train.drop("TARGET_D", axis = 1).columns)
X_train_stand
```

```python
# Standardizing X_test variables
scaler = StandardScaler()
X_test_stand = scaler.fit_transform(df_test.drop("TARGET_D", axis = 1))
X_test_stand = pd.DataFrame(X_test_stand, columns = df_test.drop("TARGET_D", axis = 1).columns)
X_test_stand
```

```python
# Gathering all "Train" variables together again
df_train = pd.concat([X_train_stand, df_train["TARGET_D"]], axis = 1)
df_train
```

```python
# Gathering all "Test" variables together again
df_test = pd.concat([X_test_stand, df_test["TARGET_D"]], axis = 1)
df_test
```

```python
# Building a Random Forest model

rf_model = RandomForestRegressor(n_estimators = 100, max_depth = 7, random_state = 42)
rf_model.fit(df_train.drop("TARGET_D", axis = 1), df_train["TARGET_D"])

print("R2 (Train data):", rf_model.score(df_train.drop("TARGET_D", axis = 1), df_train["TARGET_D"]))
print("R2 (Test data):", rf_model.score(df_test.drop("TARGET_D", axis = 1), df_test["TARGET_D"]))
```

```python
# Hyperparameter tuning 1

param_grid = {'n_estimators': [50, 100, 200],
              'max_depth': [2, 4, 6],
              'min_samples_split': [2,4],
              'min_samples_leaf': [1,2]}

rf_model_tuning = RandomForestRegressor(random_state = 42)
grid_search = GridSearchCV(rf_model_tuning, param_grid, cv = 5, return_train_score = True, n_jobs = -1 )

grid_search.fit(df_train.drop("TARGET_D", axis = 1), df_train["TARGET_D"])
print("Best parameters:")
grid_search.best_params_
```

```python
rf_model = RandomForestRegressor(n_estimators = 200, max_depth = 4, min_samples_leaf = 2, min_samples_split = 2, random_state = 42)
rf_model.fit(df_train.drop("TARGET_D", axis = 1), df_train["TARGET_D"])

print("R2 (Train data):", rf_model.score(df_train.drop("TARGET_D", axis = 1), df_train["TARGET_D"]))
print("R2 (Test data):", rf_model.score(df_test.drop("TARGET_D", axis = 1), df_test["TARGET_D"]))
```

```python
# Hyperparameter tuning 2

param_grid = {'n_estimators': [200, 250, 300, 315],
              'max_depth': [3,4,5,6,7]}

rf_model_tuning = RandomForestRegressor(random_state = 42)
grid_search = GridSearchCV(rf_model_tuning, param_grid, cv = 10, return_train_score = True, n_jobs = -1 )

grid_search.fit(df_train.drop("TARGET_D", axis = 1), df_train["TARGET_D"])
print("Best parameters:")
grid_search.best_params_
```

```python
rf_model = RandomForestRegressor(n_estimators = 300, max_depth = 3, random_state = 42)
rf_model.fit(df_train.drop("TARGET_D", axis = 1), df_train["TARGET_D"])

print("R2 (Train data):", rf_model.score(df_train.drop("TARGET_D", axis = 1), df_train["TARGET_D"]))
print("R2 (Test data):", rf_model.score(df_test.drop("TARGET_D", axis = 1), df_test["TARGET_D"]))
```

```python
# Trying a linear model to check the differences
lm_model = linear_model.LinearRegression()
lm_model.fit(df_train.drop("TARGET_D", axis = 1), df_train["TARGET_D"])

print("R2 (Train data):", lm_model.score(df_train.drop("TARGET_D", axis = 1), df_train["TARGET_D"]))
print("R2 (Test data):", lm_model.score(df_test.drop("TARGET_D", axis = 1), df_test["TARGET_D"]))
```
