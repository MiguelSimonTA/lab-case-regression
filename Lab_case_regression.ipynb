```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_columns', None)

from sklearn.model_selection import train_test_split
```

```python
# using the dataset given by the TA
numerical = pd.read_csv(r"C:\Users\Pedro Gomes\Desktop\Filipa_Ironhack\Classes\ML\numerical.csv")
target = pd.read_csv(r"C:\Users\Pedro Gomes\Desktop\Filipa_Ironhack\Classes\ML\target.csv") 
numerical.dtypes
```

```python
# Lets check null values in percentage 

nulls_percent_df= numerical.isna().sum()/len(numerical)
nulls_percent_df

# put it in a dataframe 
nulls_percent_df= pd.DataFrame(numerical.isna().sum()/len(numerical))
nulls_percent_df

# Take out of the index 
nulls_percent_df= pd.DataFrame(numerical.isna().sum()/len(numerical)).reset_index()
nulls_percent_df

# Lets change columns name
nulls_percent_df.columns = ['column_name', 'nulls_percentage']
nulls_percent_df

# Lets sort 
nulls_percent_df.sort_values(by = ['nulls_percentage'], ascending = False)


# since there are no null values we dont have to do anything
```

```python
target = target.drop(['TARGET_B'], axis=1)
```

```python
numerical
```

```python
# define the X and y 

X = numerical
y = target

# Doing the train-test-split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)

from sklearn.ensemble import RandomForestRegressor

clf = RandomForestRegressor(n_estimators = 50, max_depth = 7, random_state = 42)
clf.fit(X_train, y_train)

print(clf.score(X_train, y_train))
print(clf.score(X_test, y_test))
```

```python
from sklearn.model_selection import GridSearchCV
#  Tunning hyperparameters

param_grid = {'n_estimators': [20, 60, 80],
              'max_depth': [2, 4, 8],
              'min_samples_split': [2,4],
              'min_samples_leaf': [1,2]}

clf = RandomForestRegressor(random_state = 42)
grid_search = GridSearchCV(clf, param_grid, cv=2, return_train_score=True, n_jobs = -1 )

grid_search.fit(X_train, y_train)
grid_search.best_params_
```

```python
clf = RandomForestRegressor(n_estimators = 60, max_depth = 2, random_state = 42, min_samples_leaf = 2, min_samples_split = 2 )
clf.fit(X_train, y_train)

print(clf.score(X_train, y_train))
print(clf.score(X_test, y_test))
```

```python
# dataframe that shows the comparison between the money donated and the predicted values
predictions = pd.DataFrame([y_test, clf.predict(X_test)], axis = 1)
```
